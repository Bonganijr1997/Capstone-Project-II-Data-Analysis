```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram to visualize the distribution of car prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# We create a histogram to understand the distribution of car prices.
# The histogram helps us identify the price range in which most cars fall
```

```python
# Create a scatter plot to explore the relationship between horsepower and car prices
plt.figure(figsize=(10, 6))
plt.scatter(df['horsepower'], df['price'], alpha=0.5, color='green')
plt.title('Horsepower vs. Car Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.grid(True)
plt.show()

#Data Story:
#We create a scatter plot to analyze how horsepower relates to car prices.
#This plot helps us identify any correlation between these two variables.
```

```python
# Create a bar plot to visualize the distribution of car body styles
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='body-style', palette='magma')
plt.title('Distribution of Car Body Styles')
plt.xlabel('Body Style')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# We create a bar plot to show the distribution of car body styles.
# This plot provides insights into the popularity of different body styles.
```
